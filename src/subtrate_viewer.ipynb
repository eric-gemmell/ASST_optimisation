```python
from PIL import Image
import csv
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
```

```python
csv_file_path = '../data/ASST_substrates/substrates.csv'
image_folder = "../data/ASST_substrates/"
```

```python
def load_csv(file_path, encoding='utf-8'):
    substrate_data = []

    with open(file_path, 'r', encoding=encoding, errors='replace') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        
        for row in csv_reader:
            substrate_data.append(row)

    return substrate_data

def load_images(substrate_data):
    fig_height = len(substrate_data) * 100  # Set the total height of the figure

    # Create a 2-column grid with variable row heights
    gs = gridspec.GridSpec(len(substrate_data), 2, height_ratios=[100] * len(substrate_data), hspace=0)

    for i, substrate in enumerate(substrate_data):
        image_path = os.path.join(image_folder, substrate["image_filename"])

        # Display image and name
        plt.figure(figsize=(10, fig_height / 100))  # Set the figure size based on the total height
        ax1 = plt.subplot(gs[i, 0])
        ax1.text(0.5, 0.5, substrate["substrate_name"] + " - " + substrate["CAS_number"], fontsize=10, ha='center', va='center')
        ax1.axis('off')

        ax2 = plt.subplot(gs[i, 1])
        img = mpimg.imread(image_path)
        imgplot = ax2.imshow(img)
        ax2.axis('off')
        print(substrate["substrate_name"],substrate["CAS_number"])
    plt.show()
    plt.savefig("file.png", bbox_inches='tight', pad_inches=0.1)
```

```python

# Load CSV file into substrate_data
substrate_data = load_csv(csv_file_path)
print([s["class"] for s in substrate_data])
load_images([s for s in substrate_data if (s["class"] == "hormones" or s["class"] == "steroids")])

```
